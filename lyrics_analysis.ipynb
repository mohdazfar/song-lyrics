```python
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.spatial.distance import pdist, squareform
from itertools import chain
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
import pandas as pd
import itertools
from wordcloud import WordCloud, STOPWORDS
```

```python
artist = 'ABBA' # All the analysis will be based on a specific artist because all songs might not give a good insight
df = pd.read_csv('D:/kaggle/songlyrics/songdata.csv')
df_filter = df[df['artist'] == artist] # Filtering dataframe by artist
df_filter['text'] = df_filter['text'].str.lower() # lowercase all the text
song_list = df_filter['song'].tolist() 
unique_artist = df['artist'].unique()
print(unique_artist)
```

```python
# Setting up lyrics of songs as a list for easy processing
lyrics_list = [row['text'] for i, row in df_filter.iterrows()]
```

```python
# documents to matrix using tf-idf scores
tfidf_vectorizer = TfidfVectorizer() # scikit-learn tf-idf vectorizer 
tfidf_X = tfidf_vectorizer.fit_transform(lyrics_list)
tfidf_X = tfidf_X.toarray() # Converting the object to array
```

```python
# Descriptive statistics songs from an artist
lyrics_length_list = [len(x.split(' ')) for x in lyrics_list] 
mean_length = sum(lyrics_length_list)/len(lyrics_length_list) # Mean of words in a song
min_length = min(lyrics_length_list) # Min words in a song
max_length = max(lyrics_length_list) # Max words in a song

print('Average length of songs: {} words'.format(round(mean_length, 0)))
print('Minimum length of a song: {} words'.format(min_length))
print('Maximum length of a song: {} words'.format(max_length))
```

```python
# Cosine similarity of songs by the same singer
similar_songs = {}
cosine_dist = squareform(pdist(tfidf_X, 'cosine'))
for song in cosine_dist:
    res = itertools.compress(song_list, map(lambda x: x < 0.4, song)) # Considering simlarity < 0.4
    res = [_ for _ in res ]
    if len(res) > 1:
        res = tuple(res)
        similar_songs[hash(res)] = res # Hashing a list to avoid duplicate results

print(similar_songs.values()) # Each tuple is showing similar songs with cosine similarity < 0.4
```

```python
# Convert songs to list of tokens
def song_to_token(text):
    return ' '.join(text.split()).replace('\n', '').split(' ')
```

```python
# Top 20 Most used words by the specific singer in their songs
cachedStopWords = stopwords.words("english")
all_words = []
for song in lyrics_list:
    text = ' '.join([word for word in song.split() if word not in cachedStopWords]) # removing stopwords
    tokens = song_to_token(text)
    all_words.append(tokens)

all_words = list(chain.from_iterable(all_words)) # flat list
counts = pd.Series(all_words).value_counts()[:20] # taking care of top 20
counts.plot.barh(width=0.9, color='blue')
plt.xlabel('Frequency')
plt.show()
```

```python
# TOP 20 bigrams/trigrams used by the artist
# Reason for trigram analysis is that in songs singers often use 2-words together
# for no big reason. So trigram would give better insight in this case.
from nltk.util import ngrams

trigramfdist = []
special_chars = '"(),-.'
for line in lyrics_list:
    line = ''.join(c for c in line if c not in special_chars)
    tokens = song_to_token(line)
    bigrams = ngrams(tokens, 3) # 3=trigram, 2=bigram 
    trigramfdist.append(list(bigrams))


bigramfdist = list(chain.from_iterable(trigramfdist))
concatenate_bigrams = [x[0] + ' ' + x[1] + ' ' + x[2] for x in bigramfdist] # REcombining into one sentence
counts = pd.Series(concatenate_bigrams).value_counts()[:20]
counts.plot.barh(width=0.9, color='red')
plt.xlabel('Frequency')
plt.show()
```

```python
# Showing word cloud of the specific singer
def show_wordcloud(data, title = None):
    stopwords = set(STOPWORDS)
    wordcloud = WordCloud(
        background_color='white',
        stopwords=stopwords,
        max_words=200,
        max_font_size=40,
        scale=3,
        random_state=1 # chosen at random by flipping a coin; it was heads
    ).generate(str(data))

    fig = plt.figure(1, figsize=(12, 12))
    plt.axis('off')
    if title:
        fig.suptitle(title, fontsize=20)
        fig.subplots_adjust(top=2.3)

    plt.imshow(wordcloud)
    plt.show()

all_text = ' '.join(lyrics_list)
show_wordcloud(all_text)
```
